```python
catalog = {
    "B001": {"title": "Introduction to Algorithms", "stock": 3, "daily_fine": 5},
    "B002": {"title": "Operating System Concepts", "stock": 1, "daily_fine": 5},
    "B003": {"title": "The Great Gatsby", "stock": 5, "daily_fine": 2},
    "B004": {"title": "IEEE Journal of Robotics", "stock": 2, "daily_fine": 10}
}

students = {
    "S101": {"name": "Ananya", "borrowed": 1, "max_limit": 3, "pending_fines": 0},
    "S102": {"name": "Vikram", "borrowed": 3, "max_limit": 3, "pending_fines": 15},
    "S103": {"name": "Rohan", "borrowed": 0, "max_limit": 3, "pending_fines": 0}
}

WELCOME_MSG = """
📚 ===== Central College Library Chatbot ===== 📚
Available Commands:
1. hi / hello
2. catalog (View all books)
3. status <student_id> (Check borrowed count and fines)
4. borrow <student_id> <book_id>
5. return <student_id> <book_id> <days_late>
6. pay <student_id> <amount>
7. help
8. exit
"""

def show_catalog():
    print("\n--- Library Catalog ---")
    for b_id, info in catalog.items():
        print(f"[{b_id}] {info['title']} - Copies: {info['stock']} (Fine/Day: Rs.{info['daily_fine']})")
    print("-----------------------")

def calculate_fine(book_id, days_late):
    """Mathematical calculation for late returns."""
    return catalog[book_id]["daily_fine"] * days_late

def library_chatbot(user_input):
    user_input = user_input.strip().lower()
    parts = user_input.split()
    command = parts[0] if parts else ""


    if command in ["hi", "hello"]:
        return "Hello! I am your Library Assistant. How can I help you with your reading today?"

    elif command == "help":
        return "You can type 'catalog' to see books, or use 'borrow', 'return', 'status', and 'pay' commands."

    elif command == "catalog":
        return "Type 'catalog' to view the books."


    elif command == "status":
        if len(parts) != 2:
            return "Format error. Use: status <student_id> (e.g., status S101)"

        s_id = parts[1].upper()
        if s_id in students:
            student = students[s_id]
            return f"Student: {student['name']} | Books Borrowed: {student['borrowed']}/{student['max_limit']} | Pending Fines: Rs.{student['pending_fines']}"
        else:
            return "Error: Student ID not found in database."


    elif command == "borrow":
        if len(parts) != 3:
            return "Format error. Use: borrow <student_id> <book_id> (e.g., borrow S101 B001)"

        s_id, b_id = parts[1].upper(), parts[2].upper()

        if s_id not in students:
            return "Error: Student ID not found."
        elif b_id not in catalog:
            return "Error: Book ID not found in catalog."
        elif students[s_id]["pending_fines"] > 0:
            return f"Issue denied. {students[s_id]['name']} must clear pending fines of Rs.{students[s_id]['pending_fines']} first."
        elif students[s_id]["borrowed"] >= students[s_id]["max_limit"]:
            return f"Issue denied. {students[s_id]['name']} has reached the borrowing limit of {students[s_id]['max_limit']} books."
        elif catalog[b_id]["stock"] <= 0:
            return f"Sorry, '{catalog[b_id]['title']}' is currently out of stock."
        else:

            catalog[b_id]["stock"] -= 1
            students[s_id]["borrowed"] += 1
            return f"Success! '{catalog[b_id]['title']}' issued to {students[s_id]['name']}."


    elif command == "return":
        if len(parts) != 4:
            return "Format error. Use: return <student_id> <book_id> <days_late>"

        s_id, b_id = parts[1].upper(), parts[2].upper()

        if s_id not in students:
            return "Error: Student ID not found."
        if b_id not in catalog:
            return "Error: Book ID not found."

        try:

            days_late = int(parts[3])

            if days_late < 0:
                return "Error: Days late cannot be a negative number."


            catalog[b_id]["stock"] += 1
            students[s_id]["borrowed"] = max(0, students[s_id]["borrowed"] - 1)

            if days_late > 0:

                fine = calculate_fine(b_id, days_late)
                students[s_id]["pending_fines"] += fine
                return f"Book returned. {days_late} days late. A fine of Rs.{fine} has been added to {students[s_id]['name']}'s account."
            else:
                return f"Book returned on time by {students[s_id]['name']}. Thank you!"

        except ValueError:
            return "Input Error: 'days_late' must be a whole number (e.g., 0, 2, 5)."
        except Exception as e:
            return f"An unexpected error occurred: {e}"


    elif command == "pay":
        if len(parts) != 3:
            return "Format error. Use: pay <student_id> <amount>"

        s_id = parts[1].upper()

        if s_id not in students:
            return "Error: Student ID not found."

        try:

            amount = float(parts[2])

            if amount <= 0:
                return "Error: Payment amount must be greater than zero."

            current_fine = students[s_id]["pending_fines"]

            if current_fine == 0:
                return f"{students[s_id]['name']} has no pending fines to pay."
            elif amount > current_fine:
                return f"Error: Attempting to overpay. The pending fine is only Rs.{current_fine}."
            else:

                students[s_id]["pending_fines"] -= amount
                return f"Payment of Rs.{amount} accepted. Remaining fine for {students[s_id]['name']} is Rs.{students[s_id]['pending_fines']}."

        except ValueError:
            return "Input Error: Payment amount must be a number."

    elif command == "exit":
        return "Closing library assistant. Have a great day of studying!"

    else:
        return "Invalid command. Type 'help' to see the list of valid commands."


if __name__ == "__main__":
    print(WELCOME_MSG)

    while True:
        user_input = input("\nYou: ")

        if user_input.strip().lower() == "catalog":
            show_catalog()
            continue

        response = library_chatbot(user_input)
        print("Bot:", response)

        if user_input.strip().lower() == "exit":
            break
```
